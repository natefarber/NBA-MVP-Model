# Player Statistics Data Cleaning

First we need to clean the data for our player statistics

```python
#Load csv file
stats_df=pd.read_csv('stats_df.csv')
```

```python
stats_df.head(5)
```

First we will remove the Unamed: 0 and Rk column, as we do not need them.

```python
del stats_df['Unnamed: 0']
del stats_df['Rk']
```

```python
stats_df.head(5)
```

You may have noticed that Kareem Abdul-Jabbar has an * next to his name. This means that he was an all-star for the season. However, since we want to merge this we our MVP data frame we need to remove this so we can merge the data. Sorry Kareem.

```python
#Remove *
stats_df['Player'] = stats_df['Player'].str.rstrip('*')
```

```python
stats_df.head(5)
```

The asterisk is now gone!

If you explored the data on the basketball reference website you may have seen that some players names are listed twice in a given year. This is because they played for more than one team during that season. We need to fix this so they are only listed once. In case you did not look at the basketball reference website. I've defined a function to show that some players appear twice.

```python
def player_count_year(df, year):
    #Filter data frame
    filtered_df = df[df['Year'] == year]
    # Count the occurrences of each player
    player_counts = filtered_df['Player'].value_counts()
    
    return player_counts
```

```python
#Call the function on a year
player_count_year(stats_df, 1987)
```

We can see that there are multiple players that appear twice (or more) in the data set. First let's look closer at Chris Engler playing for three teams in 1987.

```python
stats_df.groupby(['Player', 'Year']).get_group(('Chris Engler', 1987))
```

To deal with this we a will define a function the converts players with multiple rows, like Chris Engler, to a single row. Luckily, basketball reference has a row called TOT which is their total stats for the season so we do not have to deal with calculating this ourselves. TOT is not a valid NBA team so we will replace TOT with the name of the players most recent team for the season.

```python
def one_team(df):
    return df if df.shape[0] == 1 else df[df["Tm"].eq("TOT")].assign(Tm=df.iloc[-1]["Tm"])

stats_df = stats_df.groupby(["Player", "Year"]).apply(one_team)
```

```python
stats_df.head()
```

```python
#Check to see if the function worked
player_count_year(stats_df, 1987)
```

Since every player is only listed once now, we can see that our function worked. However, we can see that the indexes of our dataframe have changed. we need to correct this.

```python
#Fix the indexes
stats_df.index = stats_df.index.droplevel()
stats_df.index = stats_df.index.droplevel()
```

```python
stats_df.head(5)
```

# MVP Data Cleaning

Now, we will clearn the data for the mvp_df file

```python
#Load csv file
mvp_df=pd.read_csv('mvp_df.csv')
```

```python
mvp_df.head(5)
```

There are some columns here that we will need to get rid of. For example, AST, STL, BLK and so on, are all apart of the player stats dataset so we do not need them here. Furthermore, the Unnamed: 0 column is not needed.

```python
#Select the columns of interest out of mvp_df
mvp_df = mvp_df[['Player', 'Year', 'Pts Won', 'Pts Max', 'Share']]
```

```python
mvp_df.head(5)
```

Now our dataframe is reduced to just show the key columns. Let's merge this dataframe with the stats_df.

```python
merged_df = stats_df.merge(mvp_df, how='outer', on=['Player', 'Year'])
```

```python
merged_df.head(5)
```

Looks good! Lets just check the data on a player who recieved NBA votes in their career.

```python
merged_df[merged_df['Player'] == 'LeBron James'].head(5)
```

Nice, looks like it worked for players who recieved votes. We need to make it so the NaN values are 0 in Pts Won and Share because this means the player recieved no votes. We also need to make the Pts Max equal to the given max for that year.

```python
merged_df[['Pts Won', 'Share']] = merged_df[['Pts Won', 'Share']].fillna(0)
```

```python
merged_df.head(5)
```

Now we need to fill the Pts Max column. The Pts Max is different in each year so we need to find its value.

```python
max_pts_by_year = merged_df.groupby('Year')['Pts Max'].max()
max_pts_by_year = pd.DataFrame(max_pts_by_year)
max_pts_by_year = max_pts_by_year.reset_index()
max_pts_by_year.head(5)
```

Now we need to fill the NaN values with the corresponding Pts Max

```python
#Fill NaN values
merged_df['Pts Max'] = merged_df['Pts Max'].combine_first(merged_df['Year'].map(max_pts_by_year.set_index('Year')['Pts Max']))
```

```python
merged_df.head(5)
```

# Team Record Data Cleaning

```python
teams_df = pd.read_csv('team_record_df.csv')
```

```python
teams_df.tail(15)
```

You can see that we have some rows where the data is messed up, like in row 1086 and 1087. We need to delete these rows.

```python
#Drop the rows with NaN values for the Overall column
teams_df = teams_df.dropna(subset=['Overall'])
#Drop the rows that don't contain '-' as this is used for all of the record
teams_df = teams_df[teams_df['Overall'].str.contains('-', na=False)]
```

```python
teams_df.tail(15)
```

Much better!

```python
#Keep only the columns of interests
teams_df = teams_df[['Team', 'Overall', 'Year']]
```

```python
teams_df.head(5)
```

Now we need to deveop a colums for wins, losses, and win percentage. We will use the Overall column to do this.

```python
#Wins column
teams_df['Wins'] = teams_df['Overall'].apply(lambda x: int(x.split('-')[0]))
#Losses column
teams_df['Losses'] = teams_df['Overall'].apply(lambda x: int(x.split('-')[1]))
#Win percentage column
teams_df['W%'] = round(teams_df['Wins'] / (teams_df['Wins'] + teams_df['Losses']), 5)
```

```python
teams_df.head(5)
```

The team names in this data frame do not match with those in the merged_df. We can see this below.

```python
teams_df['Team'].value_counts()
```

```python
merged_df['Tm'].value_counts()
```

This a problem because we are going to want to match these values so we will need to find a way to correct this. To do this I will use a file that cointains the name of NBA teams and their common abbreviations. Both dataframes have the same number of unique items so it is likely that we have an abbreviation for each team name.

```python
#Read file
nicknames = pd.read_csv('nicknames.csv')
```

```python
nicknames.head(5)
```

```python
nicknames[['Abbreviation', 'Name']] = nicknames['Abbreviation,Name'].str.split(',', n=1, expand=True)
del nicknames['Abbreviation,Name']
```

```python
nicknames.head(5)
```

Now we can use this to add full team name to our merged dataframe.

```python
#Map
merged_df['Team'] = merged_df['Tm'].map(nicknames.set_index('Abbreviation')['Name'])
```

```python
merged_df.head(10)
```

Finally, we need to combines out team_df with merged_df to get a final dataframe containing all stats.

```python
#Merge to all_stats
all_stats = merged_df.merge(teams_df, how='outer', on=['Team', 'Year'])
```

```python
all_stats.head(5)
```

```python
all_stats.columns
```

```python
#Delete overall columns since we have wins and losses
del all_stats['Overall']
```

```python
all_stats.dtypes
```

```python
#Conver final data set to a dataframe
all_stats.to_csv('all_stats.csv')
```
