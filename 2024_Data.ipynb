# 2024 Data

Now that the 2024 NBA regular season has concluded, I will be using the model to predict who the MVP will be. This notebook will be used to gather the data. Since this is only one year's worth of data, I will not be using web scraping to get the data, as the data can easily be downloaded as a csv file from Basketball Reference.

```python
#Import libraries
import pandas as pd
```

# Player Data

```python
#Import and view data
playerData = pd.read_csv('/Users/natefarber/Desktop/Projects/MVP Project/PlayerStats.csv')
playerData.head(5)
```

```python
#Delete unwanted columns
del playerData['Rk']
del playerData['Player-additional']
```

```python
#Remove * if any
playerData['Player'] = playerData['Player'].str.rstrip('*')
```

```python
playerData['Player'].value_counts()
```

```python
#Fix duplicate players (players traded mid season)
playerData = playerData.groupby("Player").apply(lambda df: df if df.shape[0] == 1 else df[df["Tm"].eq("TOT")].assign(Tm=df.iloc[-1]["Tm"]))
```

```python
playerData['Player'].value_counts()
```

```python
playerData.head(5)
```

```python
#Fix index and view
playerData.index = playerData.index.droplevel()
playerData.head(5)
```

# Team Stats

```python
#Import and view DF
teamData = pd.read_csv('/Users/natefarber/Desktop/Projects/MVP Project/teamStats.csv', header=0, skiprows=1)
teamData['Year'] = 2024
teamData.head(30)
```

```python
#Filter to columns of interest
teamData = teamData[['Team', 'Overall', 'Year']]
teamData.head(5)
```

```python
#Wins column
teamData['Wins'] = teamData['Overall'].apply(lambda x: int(x.split('-')[0]))
#Losses column
teamData['Losses'] = teamData['Overall'].apply(lambda x: int(x.split('-')[1]))
#Win percentage column
teamData['W%'] = round(teamData['Wins'] / (teamData['Wins'] + teamData['Losses']), 5)
```

```python
teamData.head(5)
```

```python
#Read in team nicknames
nicknames = pd.read_csv('nicknames.csv')
```

```python
nicknames[['Abbreviation', 'Name']] = nicknames['Abbreviation,Name'].str.split(',', n=1, expand=True)
del nicknames['Abbreviation,Name']
nicknames.head(5)
```

```python
#Change team names
playerData['Team'] = playerData['Tm'].map(nicknames.set_index('Abbreviation')['Name'])
```

```python
playerData.head(5)
```

```python
#Merge player and team stats
stats2024 = playerData.merge(teamData, how='outer', on=['Team'])
stats2024.head(5)
```

```python
stats2024.columns
```

```python
del stats2024['Overall']
```

```python
#Check for missing values
stats2024.isna().sum().sort_values(ascending=False)
```

```python
#Replace missing values with position medians
columns_to_fill = ['3P%', 'FT%', '2P%', 'eFG%', 'FG%']
for column in columns_to_fill:
    median_by_position_year = stats2024.groupby(['Year', 'Pos'])[column].transform('median') 
    stats2024[column].fillna(median_by_position_year, inplace=True)  
    
#Check for missing values
stats2024.isna().sum().sort_values(ascending = False)
```

Now that the missing values have been handled, I will create dummy variables for the position and team variables.

```python
# Create dummy variables for the Pos column
position_dummies = pd.get_dummies(stats2024['Pos'],)
team_dummies = pd.get_dummies(stats2024['Team'])
stats_dummies = pd.concat([stats2024, position_dummies, team_dummies], axis=1)
stats_dummies.drop('Pos', axis=1, inplace=True)
stats_dummies.drop('Team', axis=1, inplace=True)
```

```python
#Above average metric
average_stats = stats_dummies.groupby('Year')[['PTS', 'ORB', 'DRB', 'TRB', 'STL', 'AST', 'BLK']].transform('mean')
#Calculate the new columns by dividing player's statistics by average statistics for their year
statistics_columns = ['PTS', 'ORB', 'DRB', 'TRB', 'STL', 'AST', 'BLK']
for col in statistics_columns:
    stats_dummies[f'{col}_C'] = stats_dummies[col] / stats_dummies[col]
```

```python
stats_dummies.columns
```

PF appears twice for personal fouls and as a dummy variable for power forward. Below, I will change personal fouls per game to fouls per game (FPG).

```python
#Column Names
column_names = list(stats_dummies.columns)

#Get index
pf_index = column_names.index("PF")

#Rename first PF
column_names[pf_index] = "FPG"

#Rename columns
stats_dummies.columns = column_names

#Print updated column names
stats_dummies.columns
```
